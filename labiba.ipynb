```python
import pandas as pd
data = pd.read_csv("diamonds.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
%config InlineBackend.figure_format = 'retina'
plt.style.use("seaborn")
```

```python
import warnings
warnings.filterwarnings("ignore")
```

Histogram of number of diamonds corresponding to its length.

```python
plt.hist(data['x'], bins= 10, color='red')
plt.xlabel ("Length of a diamond")
plt.ylabel ("Count")
plt.title("Histogram of length of diamonds")
plt.show()
```

We notice that for the lowest lengths, there aren't any diamonds. However, when the length increases, the number of diamonds decreases progressively.

### Two variable plots

#### Scatterplot of price of diamonds according to their length

```python
plt.scatter(data['x'], data['price'])
plt.xlabel("Length of a diamond")
plt.ylabel("Price of diamond")
plt.title("Scatter Plot of length and price")
plt.show();
```

From this scatterplot, we notice that as the length of the diamond increases, so does its price.

### Three variable plot

#### Scatter plot of price of diamonds depending on its length and categorised by color
As the length of diamond increases its price increases as well,price is not much impacted by the color of the diamond.

```python
sns.scatterplot(data['x'], data['price'], hue = data['color'])
plt.title("Scatterplot of price by length and color")
plt.xlabel("Length of a diamond")
plt.ylabel("Price of diamond")
plt.show();
```
